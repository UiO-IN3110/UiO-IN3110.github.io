# Visualising map data with ipyleaflet

There are tools like cartopy and ipyleaflet that let us.

GeoJSON is a format for encoding geospatial data.

We are going to look at a map 

```python
import requests

base_url = (
    "https://github.com/smistad/konverter-norgeskart-projeksjon/releases/download/v2020"
)

fylker_url = f"{base_url}/fylker_komprimert.json"
kommune_url = f"{base_url}/kommuner_komprimert.json"


fylker_data = requests.get(fylker_url).json()
```

```python
fylker_data["features"][0]
```

Add single string 'name' property so it works with search later

```python
for feature in fylker_data["features"]:
    feature["properties"]["name"] = feature["properties"]["navn"][0]["navn"]
```

```python
import json
import os

import requests
from ipyleaflet import AwesomeIcon, GeoJSON, LayerGroup, Map, Marker, SearchControl

m = Map(zoom=4, center=[65, 10])
m
```

```python
fylker = GeoJSON(data=fylker_data)

m.add_control(
    SearchControl(
        position="topleft",
        layer=LayerGroup(layers=(fylker,)),
        zoom=4,
        property_name="name",
        marker=Marker(
            icon=AwesomeIcon(name="check", marker_color="green", icon_color="white")
        ),
    )
)
```
